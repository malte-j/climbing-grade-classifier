```python
import get_processed_data
import matplotlib.pyplot as plt
```

```python
# Collect acceleration magnitudes for each grade
data = {}

col = "#97C900"

# use inter font
plt.rcParams['font.family'] = 'Inter'

for grade, measurements in get_processed_data.processed_data_malte.items():
    data[grade] = []
    for measurement in measurements:
        # calculate the acceleration magnitude
        acc_mag = ((measurement["FreeAcc_X"]**2 + measurement["FreeAcc_Y"]**2 + measurement["FreeAcc_Z"]**2)**0.5).mean()
        data[grade].append(acc_mag)

# set colors to col
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=[col])



#remove top and right borders
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Create boxplot
plt.boxplot(data.values(), labels=data.keys())

# Add labels
plt.ylabel("Acceleration Magnitude")

plt.xlabel("Grade")

plt.savefig('exports/acc_boxplot.png', dpi=300)

plt.show()
```
